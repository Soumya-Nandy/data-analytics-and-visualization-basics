Generating Data

```python
from numpy.random import randn, randint, uniform, sample
```

```python
df=pd.DataFrame(randn(1000), index=pd.date_range('2019-06-07', periods=1000),columns=['value']) 
#inciment date by 1 day
```

```python
df
```

```python
ts=pd.Series(randn(1000), index=pd.date_range('2019-06-07', periods=1000))
#generating random values against date
```

```python
ts
```

```python
df['value']=df['value'].cumsum()  #cumsum() stands for cumulative sum
df.head()
```

```python
ts=ts.cumsum()  #in time series we need not to pass column name
ts.head()
```

```python
#Line Plot
ts.plot()
```

```python
#increase the size of the plot
ts.plot(figsize=(10,5))
```
