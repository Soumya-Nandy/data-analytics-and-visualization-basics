```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
```

```python
from plotly.offline import iplot
import plotly as py
```

```python
import cufflinks as cf
```

```python
py.__version__
```

```python
py.offline.init_notebook_mode(connected=True)
cf.go_offline()
```

### Line Plot

```python
df=pd.DataFrame(np.random.randn(100,3), columns=['a','b','c'])
df.head()
```

```python
df=df.cumsum()
df.head()
```

```python
df.shape
```

```python
df.iplot()
```

### Scatter Plot

```python
df.iplot(x='a',y='b', mode='markers', size=25)
```

```python
df.iplot(x='a',y='b', mode='markers', size=25, title='Scatter Plot', xTitle='X', yTitle='Y')
```

### Bar Plot

```python
titanic=sns.load_dataset('titanic')
titanic.head()
```

```python
titanic.shape
```

```python
titanic.iplot(x='sex', y='survived', kind='bar', title='Titanic')
```

```python
df.iplot(kind='bar')
```

```python
df.iplot(kind='bar', barmode='stack', bargap=0.5)
```

```python
df.iplot(kind='barh', barmode='stack', bargap=0.5)
```

### Box Plot & Area Plot

```python
df.iplot(kind='box')
```

```python
df.iplot(kind='area')
```

```python
df.iplot(kind='area', fill=True)
```

### 3D Plot

```python
df.iplot(kind='surface')
```

```python
df.iplot(kind='surface', colorscale='rdylbu')
```

```python
cf.datagen.sinwave(10,0.25).iplot(kind='surface')
```

### Spread Plot & Hist Plot

```python
df.iplot(kind='spread')
```

```python
df.iplot(kind='hist', bins=25, barmode='overlay', bargap=0.5)
```

### Bubble Plot & Heatmap

```python
bb=cf.datagen.bubble3d(5,4, mode='stocks')
bb.head()
```

```python
bb.iplot(kind='bubble3d', x='x', y='y', z='z', size='size')
```

```python
df.corr()
```

```python
df.corr().iplot(kind='heatmap', colorscale='spectral')
```

```python
cf.datagen.heatmap(20,20).iplot(kind='heatmap')
```
